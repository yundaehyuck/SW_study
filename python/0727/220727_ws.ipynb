# 7-4. seeding the generator

컴퓨터 프로그래밍에서 시드를 설정하면 동일한 순서로 난수를 발생시킬 수 있다

7-4-1) 아래의 코드를 실행하여 #1, #2에 출력되는 결과를 작성하고 seed()는 어떤 종류의 메소드인지 작성  


**답) seed를 설정했기때문에 몇번을 반복해도 #1은 이진호 #2는 강은주**

**Faker()라는 class를 통해 .seed() method를 호출했기때문에 classmethod이다**


제목 없음.jpg


**실제로도 코드 보면 classmethod**

```python
from faker import Faker

fake1 = Faker('ko_KR')
Faker.seed(87654321) #Faker라는 class에서 method를 호출했기때문에 classmethod

print(fake1.name()) #이진호

fake2 = Faker('ko_KR')
print(fake2.name()) #강은주
```

7-4-2) 아래의 코드를 실행할 때 #1, #2에서 출력되는 결과를 각각 작성하고 seed_instance()는 어떤 종류의 메소드인지 작성  

**답) #1은 이진호, #2는 김상호,최지우,이성진,강하윤,박주원,최정희,....... 실행할때마다 달라짐**

**fake1이라는 instance를 통해 seed_instance()가 호출되었으므로 instance method이다. 실제로 다른 instance인 fake2에는 seed가 동작하고 있지 않다**

제목 없음.jpg

**실제 코드도 instance method**

```python
fake1 = Faker('ko_KR')

fake1.seed_instance(87654321) #fake1이라는 instance를 통해 호출했고 instance인 fake1에만 seed가 작동하고 다른 instance인 fake2에는 작동하지 않으므로 instance method

print(fake1.name()) #이진호

fake2 = Faker('ko_KR')

print(fake2.name()) #김상호,최지우,이성진,강하윤,박주원,최정희,.... 실행할때마다 다름
```
